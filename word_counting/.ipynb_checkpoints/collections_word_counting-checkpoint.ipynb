```python
import pandas as pd
import re
from collections import Counter

class ProcessText():
    """
        This class is an attempt to clean and filter the words
        ocurrency in a collection.txt 
    Inputs : path_to_txt_file.txt
    Outputs : Dictionary as hash table of word ocurrency in Inputs.
    
    """
    def __init__(self, path_to_txt):
        self.filename = path_to_txt
        self.data = pd.read_fwf(filename, header=None, names=["collection"])
    def return_txt(self):
        """
            Collect the self.data as numpy array and returns
            a list of all the articles
        """
        collection = self.data['collection']
        collection = list(collection)
        return collection
    
    def get_individual_articles(self):
        """
            Inputs the list of articles and outputs three lists
            article_one, article_two, article_trhee
        """
        collection = self.return_txt()
        for index, art in enumerate(collection):
            
            if art == '<doc>':
                pass
                #print('X Article start here ', index+1)
            if art == '</doc>':
                pass
                #print('X Article ends here', index+1)
            
        # FALTA GENERALIZAR...
        article_one = collection[0:182]
        article_two = collection[182:237]
        article_three = collection[237:]

        assert((len(article_one) + len(article_two)+ len(article_three)) == len(collection))
        return article_one, article_two, article_three
    
    def get_unique_words_in_articles(self, article):
        unique_words_in_article = list(set(article))
        # Split the frases in this article in unique words
        unique_list = []
        for word in unique_words_in_article:
            w  = word.split(' ')
            unique_list.append(w)
            
        # Join all the words in lists in unique_lists in one single list
        unique_words = []
        for list_or_word in unique_list:               
            if type(list_or_word) == list:
                for word in list_or_word:
                    unique_words.append(word)
        
        unique_words = list(set(unique_words))
        vocab = self.limpiar_list(unique_words)
        
        # For the full article
        
        splited_articles =  list(article)
        splited_ = []
        
        for words in splited_articles:
            w = words.split(' ')
            splited_.append(w)
        
        # Join all the words in lists in unique_lists in one single list
        unique_article = []
        for list_or_word in splited_:               
            if type(list_or_word) == list:
                for word in list_or_word:
                    unique_article.append(word)
        
        unique_article = list(set(unique_article))
        full_arti = self.limpiar_list(unique_article)
        
        return vocab, full_arti

    
        # return a list of unique words in this article
    
    def limpiar_list(self, unique_words):
        # Reemplzae unsense words like commas and dots  amd )
        vocab = [u.replace(',"', '') for u in unique_words]
        vocab = [u.replace(",", '') for u in vocab]
        vocab = [u.replace(")", '') for u in vocab]
        vocab = [u.replace("(", '') for u in vocab]
        vocab = [u.replace(".", '') for u in vocab]
        vocab = [u.replace('"', '') for u in vocab]
        vocab = [u.replace(':', '') for u in vocab]
        vocab = [u.replace('$', '') for u in vocab]
        vocab = [u.replace(';', '') for u in vocab]

        vocab = [u.replace('rights', 'right') for u in vocab]
        vocab = [u.replace('levels', 'level') for u in vocab]
        vocab = [u.replace('events', 'event') for u in vocab]
        vocab = [u.replace("country's", 'country') for u in vocab]

        for i, w in enumerate(vocab):
            if w == '':
                vocab.pop(i)
        return vocab
    def create_lookup_tables(self, words):
        """
            Create lookup tables for vocabulary
            :param words: Input list of words
            :return: A tuple of dicts.  The first dict....
        """
        word_counts = Counter(words)
        sorted_vocab = sorted(word_counts, key=word_counts.get, reverse=True)
        int_to_vocab = {ii: word for ii, word in enumerate(sorted_vocab)}
        vocab_to_int = {word: ii for ii, word in int_to_vocab.items()}

        return vocab_to_int, int_to_vocab
    
    def create_hash_table(self, word):
        
        """
            Clean and extract the integer representation of the word and viseversa
            :parameter  -> word  <string>
            :returns -> list of word counted in the articles a1, a2, a3
        """
        a1,a2,a3 = self.get_individual_articles()

        vocab_in_one, full_art_one = self.get_unique_words_in_articles(a1)
        vocab_in_two, full_art_two = self.get_unique_words_in_articles(a2)
        vocab_in_three, full_art_three = self.get_unique_words_in_articles(a3)
        
        
        # Create vocav to intenger and vise versa
        vocab_to_int_one, int_to_vocab_one = process.create_lookup_tables(vocab_in_one)
        vocab_to_int_two, int_to_vocab_two = process.create_lookup_tables(vocab_in_two)
        vocab_to_int_three, int_to_vocab_three = process.create_lookup_tables(vocab_in_three)


        
        # Create integer representation of the articles for plot histograms
        interget_representation_for_one = [vocab_to_int_one[elem] for elem in full_art_one]
        interget_representation_for_two = [vocab_to_int_two[elem] for elem in full_art_two]
        interget_representation_for_three = [vocab_to_int_three[elem] for elem in full_art_three]
        
        words_as_integers_and_his_counts_for_one = dict(Counter(interget_representation_for_one))
        words_as_integers_and_his_counts_for_two = dict(Counter(interget_representation_for_two))
        words_as_integers_and_his_counts_for_three = dict(Counter(interget_representation_for_three))

        
        # Creatin a list of all the words as integers and his counts
        words_as_integers_and_his_counts_all = [words_as_integers_and_his_counts_for_one,\
                                                words_as_integers_and_his_counts_for_two,\
                                                words_as_integers_and_his_counts_for_three]
        
        # Create text representation of the articles for hash table access
        text_representation_for_one = [int_to_vocab_one[elem] for elem in  interget_representation_for_one]
        text_representation_for_two = [int_to_vocab_two[elem] for elem in  interget_representation_for_two]
        text_representation_for_three = [int_to_vocab_three[elem] for elem in  interget_representation_for_three]
        
        
        integers_as_words_and_his_counts_for_one = dict(Counter(text_representation_for_one))
        integers_as_words_and_his_counts_for_two = dict(Counter(text_representation_for_two))
        integers_as_words_and_his_counts_for_three = dict(Counter(text_representation_for_three))

        # Creatin a list of all the words as integers and his counts
        integers_as_words_and_his_counts_all = [integers_as_words_and_his_counts_for_one,\
                                                 integers_as_words_and_his_counts_for_two,\
                                                 integers_as_words_and_his_counts_for_three]
        
        for i, article in enumerate(integers_as_words_and_his_counts_all):
            print(word, [i+1, article[word]])
        
        #return  words_as_integers_and_his_counts_all, integers_and_words_and_his_counts_all 
        
# path where the file lives
filename = 'collection.txt'
# create the object
process = ProcessText(filename)
# get the hash table for the word 'a'
process.create_hash_table('a')
```
